```python
from langchain_anthropic import ChatAnthropic
from dotenv import load_dotenv
load_dotenv()
import os 

if os.getenv("ANTHROPIC_API_KEY") is None:
    raise ValueError("ANTHROPIC_API_KEY environment variable not set")
else:
    print("ANTHROPIC_API_KEY environment variable is set")
```

```python
llm = ChatAnthropic(model="claude-haiku-4-5-20251001")

response = llm.invoke("""Give me the code to create a pandas dataframe with 3 columns and 5 rows.
                      Please also generate a python file to store the code. Please.""")
response.content
```
